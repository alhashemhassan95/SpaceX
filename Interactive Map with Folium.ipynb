```python
import folium

# Create base map centered around Cape Canaveral
spacex_map = folium.Map(location=[28.5623, -80.5774], zoom_start=5)

# Add markers
folium.Marker(
    location=[28.5623, -80.5774],
    popup='Cape Canaveral Launch Complex 40',
    tooltip='Falcon 9 Launch Site',
    icon=folium.Icon(color='green')
).add_to(spacex_map)

folium.Marker(
    location=[9.0477, 167.7431],
    popup='Kwajalein Atoll Launch Site',
    tooltip='Falcon 1 Launch Site',
    icon=folium.Icon(color='blue')
).add_to(spacex_map)

# âœ… This line renders the map directly in Jupyter
spacex_map
```
