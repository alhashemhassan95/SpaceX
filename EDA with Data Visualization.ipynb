```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Create and save sample SpaceX data
data = {
    'mission_name': ['FalconSat', 'DemoSat', 'Trailblazer', 'RatSat', 'RazakSat', 'COTS 1', 'COTS 2', 'CRS-1', 'CRS-2', 'SES-8'],
    'launch_date': ['2006-03-24', '2007-03-21', '2008-08-03', '2008-09-28', '2009-07-14',
                    '2010-06-04', '2012-05-22', '2012-10-08', '2013-03-01', '2013-12-03'],
    'launch_site': ['Kwajalein Atoll', 'Kwajalein Atoll', 'Kwajalein Atoll', 'Kwajalein Atoll', 'Kwajalein Atoll',
                    'Cape Canaveral', 'Cape Canaveral', 'Cape Canaveral', 'Cape Canaveral', 'Cape Canaveral'],
    'rocket_name': ['Falcon 1', 'Falcon 1', 'Falcon 1', 'Falcon 1', 'Falcon 1',
                    'Falcon 9', 'Falcon 9', 'Falcon 9', 'Falcon 9', 'Falcon 9'],
    'launch_success': [0, 0, 0, 1, 0, 1, 1, 1, 1, 1]
}

df = pd.DataFrame(data)
df.to_csv('spacex_launch_data.csv', index=False)

# Step 3: Load the CSV file
df = pd.read_csv('spacex_launch_data.csv')

# Step 4: Bar Chart - Launch Success by Site
plt.figure(figsize=(8, 5))
sns.countplot(x='launch_site', hue='launch_success', data=df)
plt.title('Launch Success by Site')
plt.xlabel('Launch Site')
plt.ylabel('Number of Launches')
plt.xticks(rotation=45)
plt.legend(title='Success')
plt.tight_layout()
plt.show()

# Step 5: Line Chart - Monthly Launch Frequency
df['launch_date'] = pd.to_datetime(df['launch_date'])
df['year_month'] = df['launch_date'].dt.to_period('M')
launch_counts = df.groupby('year_month').size()

plt.figure(figsize=(10, 4))
launch_counts.plot()
plt.title('Monthly Launch Frequency')
plt.xlabel('Date')
plt.ylabel('Number of Launches')
plt.grid(True)
plt.tight_layout()
plt.show()

# Step 6: Pie Chart - Rocket Usage
rocket_counts = df['rocket_name'].value_counts()

plt.figure(figsize=(6, 6))
rocket_counts.plot.pie(autopct='%1.1f%%', startangle=140)
plt.title('Rocket Usage Distribution')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
